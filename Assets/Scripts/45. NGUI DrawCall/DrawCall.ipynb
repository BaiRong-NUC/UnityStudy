1. DrawCall: 
    - 概念: CPU向GPU发送开始渲染的命令,游戏中DrawCall数量较高会导致性能下降
    - 降低DrawCall的方法: 在UI层面上小图合大图,多个小DrawCall合并成一次大的DrawCall
    - 制作UI时降低DrawCall的技巧: 
        - 通过NGUI Panel组件上提供的DrawCall查看工具,(使用同一个图集上的小图算作一次DrawCall)
        - 注意不同图集之间的层级关系(三张图片,层级依次递增,如果三张图片使用同一个图集,则只会产生一次DrawCall;如果第二张图片使用不同图集,第一,第三张图片图集相同,会产生三次DrawCall,也就是同一个图集的图片层级之间不要有其他图集的图片)
        - 注意Label的层级关系: Label最后也会被打包成一个图集,注意事项同上
