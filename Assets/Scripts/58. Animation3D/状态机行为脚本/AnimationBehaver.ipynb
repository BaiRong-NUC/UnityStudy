1. 状态机行为脚本是特殊的脚本,继承指定的基类,可以在进入,退出,维持状态时执行相应的逻辑
eg: 播放声音,特效,判断是否落地
2. 创建步骤:
- 创建脚本,继承自StateMachineBehaviour
- 实现其中特定的方法,实现状态机行为监听
    - OnStateEnter // 当进入某个状态时调用
    - OnStateExit  // 当退出某个状态时调用
    - OnStateIK     // OnAnimatorIK执行后调用
    - OnStateMove   // OnAnimatorMove执行后调用
    - OnStateUpdate // 除第一帧,最后一帧,每个Update调用
    - OnStateMachineEnter // 当进入子状态机时调用,第一个Update中调用
    - OnStateMachineExit  // 当退出子状态机时调用,最后一个Update中调用
