1. 子状态机: 在状态机中嵌套另一个状态机，允许更复杂的状态管理。
- 主要作用: 某一个状态是由多个动作组合成的复杂动作(eg: 技能释放包括跳跃,攻击,落地等动作,可以将这三个动作放到子状态机中管理).

2. 创建子状态机:
- 在Unity的Animator Controller窗口中,右键点击空白区域 -> "Create Sub-State Machine"来创建一个子状态机.

3. 编辑子状态机: 注意子状态机和外部状态机的相互连接方式(与(Up)layerName的连接)
- StateMachine连接:(连接外部某层的状态机),当子动画机播放完毕后,会返回到外部状态机的默认状态.
- State连接:(连接外部某层的某个状态),当子动画机播放完毕后,会返回到外部状态机的指定状态.
