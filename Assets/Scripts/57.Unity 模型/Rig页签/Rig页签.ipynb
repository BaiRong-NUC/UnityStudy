1. Rig页签: 主要骨骼映射到导入模型的网格,以便其可以动画化
    - 对于人形模型需要创建或分配Avator(替身信息),对于非人性模型需要确定根骨骼
2. Rig面板参数:
    - Animtion Type:动画类型
        - Humanoid:人形模型,需要使用Avator设置人和主要的关节的映射
            - Avatar Definition:替身定义信息
                - No Avatar:无替身,不使用Avator
                - Create From This Model:从当前模型创建Avator
                    - Configure:配置Avator,打开Avator配置窗口,可以手动调整骨骼映射
                - Copy From Other Avatar:指向另一个模型的Avator
                    - Source:来源模型,选择另一个已经有Avator的模型
        - Generic:非人形模型(通用模型),不是人形,可以是任意的形状,主要需要设置骨骼根节点
            - Root Bone:根骨骼,指定动画的根节点
        - Legacy:旧版动画系统,Unity早期版本使用的动画系统,一般不选
            - Generation: 动画导入方式
                - Don't Generate:不导入
                - Store in Root(New): 导入动画,存储在根节点
                - 其余参数弃用
        - None:无动画
3. Avator(化身系统): 把一个人的动作套用到另一个人形模型,只需关节信息一一对应即可
    - Maping: 关节映射信息设置
        - Pose: 角色姿势
            - Enforce T-Pose: 强制T字姿势,双臂水平伸展
            - Reset Pose: 重置姿势,将模型恢复到默认姿势
            - Sample Bind-Pose: 绑定姿势示例
    - Muscles & Settings: 肌肉和设置,调整模型的肌肉参数
        - Muscle Groups Preview: 预览不同姿势的旋转变化
        - Pre-Mouscle Settings: 设置骨骼旋转的范围限制
        - Additional Settings: 额外设置
            - Translation DoF: 启用人形模型的移动动画,会提升性能消耗,默认关闭(除非模型中包含骨骼的动画式移动,像是骨骼拉长等)
        - Mouscle->Rest All: 重置所有肌肉设置
