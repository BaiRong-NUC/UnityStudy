1. Materials页签:在材质,纹理模型上使用,不在动画上使用
    - Materials Created Mode: Unity导入材质的方式
        - None: 不创建材质
        - Standard: 创建Unity标准材质
        - Import via Material Description: 通过材质描述文件导入材质,支持多种材质类型
    - SRGB Albedo Colors: 是否在伽马空间中使用反射率颜色,使用线性颜色的控件不开启这个选项,使用Import via Material Description模式时,这个选项无效不可用
    - Location: 定义如何访问材质和纹理
        - Use External Materials (Legacy): 将倒入的资源保持提取为外部资源
        - Use Embedded Materials: 将导入的材质保持在导入的资源中
        - Textures/Materials: 将纹理和材质导入到指定的文件夹中,如果没有需要提取的子资源,则该选项为灰色
    - Naming: 定义材质的命名规则
    - Search: 定义在使用Naming选项定义的名称,查找现有材质的规则
        - Project-Wide: 在整个项目中查找现有材质
        - Recursive Up: 从导入资源所在文件夹开始,向上递归查找现有材质
        - Current Material Folder: 在本地Materials文件夹中查找现有材质
    - EthanWhite: 重新映射的材质,Unity在导入时会尝试将材质映射到这些材质上,可以手动关联
