1. Animation动画页签:一般3D模型网格信息和动画信息分开导出
    - 基础信息设置
        - import Animation: 从此资源导入动画,如果只有这个选项说明这个模型不包含任何的动画信息
        - Bake Animation: 通过方向动力学模拟计算动画,仅用于Maya,3Dmax等软件导出的模型
        - Anim.Compression: 动画压缩选项
            - Off: 关闭动画压缩
            - Keyframe Reduction: 关键帧减少,通过删除不必要的关键帧来减小动画数据大小(仅适用于通用动画)
            - Optimal: 最优压缩,通过多种算法综合压缩动画数据,通常能获得更好的压缩效果(使用与人形动画,通用动画)
        - Animated Custom Properties: 启用后,动画中的自定义FBX属性也会被导入
        - Clips: 动画剪辑的选择列表,如果动画中包含多个动画,使用此选项将动画进行切分
    - 动画剪辑基础信息设置
        - Loop Time: 播放动画是否循环播放
            - Loop Pose: 无缝循环
            - Loop Match: 循环匹配,绿色表示可以首尾循环匹配
            - Cicle Offset: 循环偏移,用于调整循环动画的起始位置
        - Root Transform Rotation: 根位置角度
            - Bake Into Pose: 将根旋转烘焙到骨骼移动,禁用动画旋转会影响根旋转
            - Based Upon: 根旋转的基础
                - Original: 保持原文件中的根旋转
                - Body Orientation: 保持上半身朝前(仅使用与人形模型)
                - Root Node Rotation: 保持根节点的旋转(仅适用于通用类型)
                - Offset: 根旋转偏移
        - Root Transform Position(Y): 根位置Y轴(垂直方向)
            - Bake Into Pose: 将根位置Y轴烘焙到骨骼移动,禁用动画位置会影响根位置Y轴
            - Based Upon(at Start): 根位置Y轴的基础
                - Original: 保持原文件中的垂直位置
                - Center of Mass: 使用质心和根位置对齐(仅使用与人形模型)
                - Root Node Position: 使用垂直方向根位置(仅适用于通用类型)
                - Feet: 使用脚部位置与根变化位置对齐(仅适用于人形模型)
            - Offset: 根位置Y轴偏移
        - Root Transform Position(XZ): 根位置X轴和Z轴(水平方向)
            - Bake Into Pose: 将根位置X轴和Z轴烘焙到骨骼移动,禁用动画位置会影响根位置X轴和Z轴
            - Based Upon: 根位置X轴和Z轴的基础
                - Original: 保持原文件中的水平位置
                - Root Node Position: 使用水平根位置(仅适用于通用类型)
                - Center of Mass: 使用质心和根位置对齐(仅使用与人形模型)
        - Mirror: 镜像动画,只有人形模型才支持此选项
        - Additive Reference Pose: 设置附加动画层的参考姿势(没有播放动画的默认姿势)
