1. Unity中支持的模型:
    - .fbx,dae,3ds,dxf,obj等常见3D模型格式
    - 建议将模型导出为.fbx格式,兼容性最好
    - 一般导出时,人物面朝Z轴正方向,上朝Y轴正方向,X轴正方向为右侧
    - 导出规范: https://docs.unity.cn/cn/2019.4/Manual/CreatingDCCAssets.html
2. 在Unity中设置模型
    - 在Project窗口中选择模型资源,在Inspector面板中设置四个页签:
      - Model: 模型标签
      - Rig: 操作(骨骼)标签
      - Animation: 动画标签
      - Materials: 材质纹理标签
