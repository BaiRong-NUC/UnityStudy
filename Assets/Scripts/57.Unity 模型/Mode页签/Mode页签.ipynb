1. Mode页签的内容

    - 模型比例设置
    - 是否导入模型的摄像机和光源
    - 网格是否压缩

2. Mode页签参数

    - Scene场景相关
        - Scale Factor: 模型全局比例,Unity中1米对应模型中的1个单位
        - Convert Units: 将模型定义的比例转化为Unity的比例
        - Preserve Hierarchy: 保持模型的层级结构不变,始终创建一个显示预制体的根
            (一般当多个FBX文件包含同一个层级的空根对象时,eg:一个fbx文件包含骨骼和网格,另一个fbx文件包含动画,此时需勾选)
    - Meshes网格相关
        - Mesh Compression: 网格压缩,减小网格数据的内存占用,但会影响网格的精度,可以优化包的大小
        - Read/Write Enabled: 允许脚本读取或写入网格数据,如果不需要动态修改网格,建议关闭以节省内存
            - 当需要在代码中读取网格数据时需要开启
            - 需要在运行时合并网格
            - 需要使用网格碰撞器
            - 需要在运行时使用NavMesh组件来烘焙NavMesh(寻路网格)等等
    - Geometry几何体相关(设置模型的法线,切线)
        - Weld Vertices: 合并在空间中共享相同位置的顶点,前提是顶点共享相同的属性(UV,法线,切线等),可以减少顶点数量,优化渲染性能
