1. NGUI的作用:
    - 降低DrawCall数量
    - 自定义美术字体
2. 制作NGUI字体:
    - 根据字体文件,生成指定内容文字,达到降低DrawCall的目的,Type: Generated Bitmap
        - Characters: 指定需要生成的文字内容
        - OutPut: 指定生成字体保存的图片保存到的图集
        - 没有在这个字符内容中的文字不会在UI中显示
    - 使用第三方工具BitmapFont生成字体信息和图集,通过NGUI字体工具使用第三方工具生成的内容制作字体达到自定义美术字体,Type: Imported Bitmap
        - Font Data: 指定第三方工具生成的字体信息文件(eg. BMFont生成的.fnt文件)
        - Texture: 指定第三方工具生成的字体图集(eg. BMFont生成的.png文件)
        - OutPut: 指定生成字体保存的图片保存到的图集
3. Unity动态字体和NGUI字体的选择
    - 文字变化比较多使用动态字体,变化少用NGUI字体
    - 想要减少DrawCall使用NGUI字体
    - 美术字使用NGUI字体
