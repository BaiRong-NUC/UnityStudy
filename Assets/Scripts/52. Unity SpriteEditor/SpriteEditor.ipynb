# 1. SpriteEditor:
SpriteEditor是精灵图片编辑器,它主要用于编辑2D游戏开发中使用的sprite精灵图片,可以用于编辑图集中提取元素，设置精灵边框，设置九宫格，设置轴心（中心）点等等功能

# 2. 安装SpriteEditor:
Window->Package Manager,在Package Manager中搜索Sprite Editor,点击安装即可

# 3. SpriteEditor的属性:
- Name: 文件名称
- Position: 在图片中的偏移位置和宽高
- Border: 边框,用于设置九宫格的四条边
- Pivot: 轴心(中心)点位置
- Pivot Unit Mode: 轴心点单位模式
  - Pixels: 像素模式
  - Normalized: 归一化模式(0-1之间)
- Custom Pivot: 自定义轴心点位置

# 4. Custom Outline:
Custom Outline用于自定义精灵的轮廓线,可以通过添加、移动、删除点来调整轮廓线的形状,从而更精确地定义精灵的碰撞区域或可见区域。

- 默认情况下不修改都是在矩形网络上渲染,边缘外部透明区域会被渲染,浪费性能,使用自定义轮廓,可以调小透明区域,提高性能
- Snap: 将控制点贴近在最近的像素
- Outline Tolerance: 轮廓点的复杂性和准确性, 0-1值越大轮廓点越多,越准确
- Generate: 生成网络轮廓

# 5. Custom Physics Shape:(决定碰撞判断区域)
属性与Custom Outline类似,用于自定义精灵的物理碰撞形状,可以通过添加、移动、删除点来调整碰撞形状,从而更精确地定义精灵的碰撞检测区域。

# 6. Secondary Textures:(为图片添加特殊效果)
次要纹理设置,可以将其他纹理和图片做关联,从而实现一些特殊效果
