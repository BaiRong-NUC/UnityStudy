# EventSystem(UICamera)作用
1. 让摄像机渲染出来的物体能够接收NGUI的输入事件,实现鼠标,键盘操作UI
# EventSystem(UICamera)属性
1. Event Type: 事件类型,决定了脚本如何对鼠标和触屏事件进行响应来处理事件
    - 3D World
    - 3D UI
    - 2D World
    - 2D UI
    - 2D,3D的区别是碰撞器使用的是2D碰撞器还是3D碰撞器;UI和World的区别是事件处理方式是根据组件的深度属性还是摄像机与物体的距离来处理事件
2. Events go to: 事件是通过刚体还是碰撞盒传递
3. Process Events In: 事件更新进度在Update还是LateUpdate中,默认在Update中
4. Event Mask: 决定那个游戏对象层级将会接受事件
5. Debug: 是狗开启调试模式,如果开启可以打印当前交互对象,在Scene窗口可以看到的信息
6. Command Click: 苹果电脑上是否启用Command键模拟右键操作
7. Auto Hide Cursor: 当有游戏控制器或者其他输入设备,是否自动隐藏光标
8. Sticky Tooltip: 是否启用粘性提示,当鼠标悬停在UI上时,提示信息会一直显示,直到鼠标移开
9. Tooltip Delay: 停留多久出现tip
10. Long Press ToolTip: 长按出现提示
11. RayCast Range: 射线探测长度,一般不修改
12. EventSources: 接受的事件来源
    - Mouse: 鼠标事件
    - Touch: 触摸事件
    - Controller: 控制器事件
    - Keyboard: 键盘事件
13. Threshold: 调整鼠标事件的点击,拖,轻拍等行为的灵敏度
14. Axes and keys: 热键关系,一般不修改
