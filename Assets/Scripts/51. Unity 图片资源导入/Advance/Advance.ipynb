# Advance:高级纹理设置
1. Non-Power of Two Textures:非2的幂次方纹理如何处理
- None:不处理，直接使用原始纹理
- ToNearest:调整为最近的2的幂次方尺寸(PVRTC要求纹理为正方形)
- ToLarger:调整为最大尺寸大小值的2的幂次方
- ToSmaller:调整为最小尺寸大小值的2的幂次方

2. Read/Write Enabled:启用后可以在运行时读取和写入纹理数据,一般需要获取图片数据时才开启

3. Generate Mip Maps:允许生成MiniMap贴图,提高远距离纹理显示效果,但会增加内存使用
- 开启MipMap功能后，Unity会根据图片信息生成n张不同分辨率的图片,在场景中会根据离该模型的距离选择合适尺寸的图片用于渲染，提升渲染效率
- Streaming Mip Maps:流式MiniMap贴图,只在需要时加载MiniMap贴图,节省内存。通常与Minmap priority一起使用
