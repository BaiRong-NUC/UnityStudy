1. WrapMode: 平铺纹理时的方式
    - Repeat: 重复平铺
    - Clamp: 拉伸边缘像素
    - Mirror: 镜像平铺
    - MirrorOnce: 镜像平铺一次后拉伸边缘像素
    - Per-Axis: 单独控制如何再U轴V轴上包裹纹理
2. FilterMode: 进行3D拉伸时如何进行过度
    - Point: 纹理在靠近时变为块状
    - Bilinear: 纹理在靠近时变得模糊
    - Trilinear: 纹理在靠近时变得模糊,与Bilinear类似，但在不同mipmap级别之间模糊过渡
3. AnisoLevel: 以大角度查看纹理时提高纹理质量。性能消耗高
