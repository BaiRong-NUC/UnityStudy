1. 平台设置主要设置纹理最终打包时在不同平台的尺寸、格式、压缩方式,它非常的重要，因为它影响了你的包大小和读取性能方面的问题

2. 参数:
- Max Size: 纹理的最大尺寸,如果你的纹理尺寸超过了这个值,引擎会自动缩小到这个值
- Resize Algorithm: 纹理缩放算法,常用的有Bilinear(双线性,差值保留细节多)和Mitchell(米切尔,常用),当图片大于Max Size时会用到
- Format: 纹理格式,各个平台支持的格式不同,如果选择Automatic,引擎会根据平台选择最合适的格式
   - IOS:选择默认的纹理压缩设置（PVRTC）可获得更大的兼容性。如果应用程序不包含OpenGLES2支持，则可以选择一种ASTC格式。这些格式提供更好的质量和灵活性，并且压缩速度比PVRTC快。
   - Android:ETC格式是大多数Android设备的默认选择，但它不支持Alpha通道。如果需要Alpha通道，可以选择ETC2或ASTC格式，这些格式提供更好的质量和压缩效果。(需要有对应版本的OpenGL ES支持)。
   - 由于安卓设备多,标准不统一,一般会构造多个安装包eg: 构造一个以OpenGL ES3的APK,另一个以OpenGL ES2的APK,分别对应不同的设备
- Compression: 纹理压缩方式,一般选择Normal Quality(正常质量),如果对质量要求不高可以选择Low Quality(低质量)以减小包体积

- Use Crunch Compression: 是否使用Crunch压缩,这种压缩方式可以显著减小纹理的存储大小,但会增加加载时间(有损压缩)
- Spilt Alpha Channel: 是否将Alpha通道分离出来单独存储,这样可以提高某些平台的渲染性能,节省内存大小(仅支持ETC压缩方式)
- Override ETC2 fallback: 不支持ETC2格式的设备使用的替代格式
