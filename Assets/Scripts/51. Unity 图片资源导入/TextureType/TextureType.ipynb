# 纹理类型主要是设置什么:
说明这张图片的用途，比如它是用来表示颜色、法线还是高度信息等。
# 参数说明:
1. Default: 默认纹理类型，通常用于表示颜色信息。
- SRGB: 启用sRGB色彩空间，适用于颜色纹理.一般勾选
- Alpha Source: 指定如何生成纹理的Alpha通道。
  - None: 不使用Alpha通道。
  - From Gray Scale: 使用输入纹理RGB的平均值作为Alpha通道。
  - Input Texture Alpha: 使用输入纹理的Alpha通道。
- Alpha Is Transparency: 启用可以避免纹理边缘出现锯齿。
2. Normal Map: 法线贴图就是在原物体的凹凸表面的每个点上均作法线,法线就是垂直于某个点的切线的方向向量.
   - 一般情况下使用精致模型的法线贴图应用到低模模型上，可以在不增加多边形数量的情况下，表现出高模的细节。
   - Create From Grayscale: 从灰度图创建法线贴图。
      - Bumpiness: 调整凹凸感强度。
      - Filtering: 如何计算凹凸值。
3. Editor GUI and Legacy GUI: 用于编辑器界面和GUI的纹理类型。主要用与编辑器开发和GUI界面设计。
4. Sprite (2D and UI): 用于2D精灵和用户界面的纹理类型。主要用于2D游戏开发和UGUI设计。
   - Sprite Mode: 定义精灵的模式。图像提取精灵的方式
     - Single: 单一图片。
     - Multiple: 瓦片模式,如果是图集可以选择这个,可以在Sprite Editor中切割成多个精灵。
     - Polygon: 多边形模式，适用于不规则形状的精灵。
   - Pixels Per Unit: 每单位像素数，决定精灵在世界空间中的大小。
   - Mesh Type: 定义精灵的网格类型。只有Single和Multiple模式下可用。
     - Full Rect: 完整矩形网格。
     - Tight: 紧密网格，适合不规则形状的精灵。注意: 任意小于32x32的贴图,如果选择Tight也会变成Full Rect.
   - Pivot: 设置精灵的枢轴点位置。Single下可以用
   - Generate Physics Shape: 为精灵生成物理形状碰撞体。只有Single和Multiple模式下可用。
   - Sprite Editor: 打开精灵编辑器，用于切割和编辑精灵。
5. Cursor: 用于光标的纹理类型。主要用于自定义鼠标指针。
6. Cookie: 用于灯光投影的纹理类型。主要用于创建灯光投影效果。
   - Light Type: 定义灯光类型。
      - SpotLight: 聚光灯。需要边缘黑色纹理
      - Point: 点光源。需要立方体贴图。
      - Directional: 方向光。需要平铺纹理。
7. Lightmap: 用于光照贴图的纹理类型。主要用于静态光照效果。
8. Single Channel: 用于单通道纹理的类型。主要用于存储高度图、金属度图等单通道信息。
   - Channel: 选择使用的通道。
      - Alpha: 使用Alpha通道。
      - Red: 使用红色通道。
