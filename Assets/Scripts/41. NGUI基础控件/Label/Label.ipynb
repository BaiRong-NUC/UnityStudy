# label的作用:
1. NGUI中所有的文本都是由Label组件来显示的
# label的参数
1. Font: 设置Label显示的字体
2. Material: 设置Label显示的材质，一般不修改
3. Font Size: 设置Label显示的字体大小
4. Text: 文本内容
5. Modifier: 文字修饰方式,强制大写/小写
6. Overflow: 文本溢出处理方式
   - Shrink Content: 自动缩小字体以适应Label区域
   - Clamp Content: 截断超出Label区域的文本,不显示
   - Resize Freely: 根据文本内容自动调整Label的大小
   - Resize Height: 根据文本内容自动调整Label的高度,只能手动调整宽度
7. Alignment: 文本对齐方式
8. Keep Crisp: 动态字体锐化,保证字体清晰
9. Gradient: 渐变颜色效果
10. Effect: 文本特效
11. Float Spacing: 浮点数间隔调整,默认整形,配合字符间距调整使用
12. Spacing: 字符间距调整
13. Max Lines: 最大显示行数
14. BBCode: 支持富文本
15. Color Tint: 文本颜色
# 富文本
1. 支持的标签
   - [颜色16进制]颜色[-]
   - [b]加粗文本[/b]
   - [i]斜体文本[/i]
   - [u]下划线文本[/u]
   - [s]删除线文本[/s]
   - [sub]右下角显示文本[/sub]
   - [sup]右上角显示文本[/sup]
   - [url=链接地址][u]显示文本[/u][/url],需要添加一个NGUI碰撞器,添加脚本Open URL On Click (NGUI自带脚本)
