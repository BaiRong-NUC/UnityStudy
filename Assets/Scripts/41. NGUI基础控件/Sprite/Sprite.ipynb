# Sprite组件属性
1. Atlas: 选择一个图集资源。
2. Sprite: 选择使用图集中的精灵图片
3. Material: 材质,一般不修改
4. Fixed Aspect: 固定宽高比,不管如何变化尺寸,都保持原始宽高比
5. Type: 图片平铺类型
    - Simple: 简单模式,不重复绘制,缩放会拉伸
    - Sliced: 切片模式,适用于9宫格图片.保证图片的四角不会发生变化,需要配合图片切片
    - Tiled: 平铺模式,图片重复绘制
    - Filled: 填充模式,可以做CD,进度条等效果
    - Advanced: 高级模式,将图片分成5个部分,分别设置这5个部分的模式
6. Flip: 反转模式(水平翻转,竖直反转,同时反转)
7. Gradient: 渐变色设置
8. Color Tint: 颜色叠加
9. Widget: 组件位置
