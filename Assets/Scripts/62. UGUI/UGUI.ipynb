1. UGUI六大基础组件
    - Canvas: 画布组件，主要用于渲染UI控件
    - Canvas Scaler: 画布分辨率自适应组件，主要用于分辨率自适应
    - Graphic Raycaster: 射线事件交互组件，主要用于控制射线响应相关
    - Rect Transform: UI对象位置锚点控制组件，主要用于控制位置和对其方式
    - Event System: 玩家输入事件响应系统
    - Standalone Input Module: 独立输入模块组件，主要用于监听玩家操作
2. Canvas: 如果UI控件对象不是Canvas的子对象, 那么控件将不能被渲染, 可以通过修改Canvas组件上的参数修改渲染方式。场景中允许有多个Canvas对象,可以分别管理不同画布的渲染方式，分辨率适应方式等等参数
    - Render Mode: 渲染模式
        - Screen Space - Overlay: 屏幕空间-覆盖模式, UI控件渲染在屏幕最前端
            - Pixel Perfect: 像素完美模式, 使UI控件在不同分辨率下都能保持清晰
            - Sort Order: 渲染顺序, 数值越大覆盖在上层
            - Target Display: 目标显示器, 多显示器时选择渲染到哪个显示器
            - Additional Shader Channels: 额外着色器通道, 控制着色器可以读取那些数据
        - Screen Space - Camera: 屏幕空间-摄像机模式, 3D物体可以显示在UI控件前面
            - Render Camera: 渲染摄像机, 指定用于渲染UI控件的摄像机(如果不设置,就类似覆盖模式),一般单独创建一个摄像机用于渲染UI控件
            - Plane Distance: 平面距离, 控制UI控件距离摄像机的远近,类似Z轴, UI上的3D物体和场景的3D物体不是一个概念
            - Sort Layer: 渲染层级, 控制UI控件的渲染层级，层级越高覆盖在上层
            - Order in Layer: 当处于同一层级时, 数值越大覆盖在上层,高层级一定覆盖低层级,无论Order in Layer数值大小
        - World Space: 世界空间模式, UI控件渲染在3D世界中, eg: UI围着3D物体旋转
            - Event Camera: 事件摄像机, 指定用于UI控件交互的摄像机,如果不设置,不能注册UI事件,一般设置为主摄像机
3. Canvas Scaler: 主要用于分辨率自适应,不负责UI控件大小适应,不负责位置适应(RectTransform负责位置适应)
    - Rect Transform中的宽高×Rect Transform中的缩放比例 = 屏幕分辨率
    - UI Scale Mode: UI适配模式
        - Constant Pixel Size: 常量像素大小, UI控件像素大小不变,适合分辨率变化不大的项目
            - Scale Factor: 缩放因子, 控制UI控件的整体缩放比例
            - Reference Pixels Per Unit: 每单位参考像素, 多少个像素等于Unity一个单位,一般为100
            - UI尺寸 = 图片像素大小 / (Pixels Per Unit(图片属性) / Reference Pixels Per Unit)
        - Scale With Screen Size: 随屏幕尺寸缩放, UI控件根据屏幕分辨率进行缩放,适合分辨率变化较大的项目
            - Reference Resolution: 参考分辨率(美术设计时的分辨率,缩放模式下的所有匹配模式都会基于参考分辨率进行自适应计算)
            - Screen Match Mode:屏幕匹配模式，当前屏幕分辨率宽高比不适应参考分辨率时，用于分辨率大小自适应的匹配模式
                - Expand: 水平或垂直拓展画布区域，会根据宽高比的变化来放大缩小画布，可能有黑边(分辨率大于参考分辨率,所有的内容都能显示出来)
                    - 缩放系数=Mathf.Min(屏幕宽/参考分辨率宽,屏幕高/参考分辨率高)
                    - 画布尺寸=屏幕尺寸/缩放系数
                - Shrink: 水平或垂直裁剪画布区域，会根据宽高比的变化来放大缩小画布，可能会裁剪(分辨率小于参考分辨率,部分内容显示不出来)
                    - 缩放系数=Mathf.Max(屏幕宽/参考分辨率宽,屏幕高/参考分辨率高)
                    - 画布尺寸=屏幕尺寸/缩放系数
                - Match Width Or Height: 以宽高或者二者的平均值作为参考来缩放画布区域,一般不存在横竖屏切换时使用
                    - Match: 匹配值, 0表示完全匹配宽度,1表示完全匹配高度,0.5表示宽高各占一半(竖屏设备建议设置为1,横屏设备建议设置为0)
        - Constant Physical Size: 常量物理大小, UI控件根据物理尺寸进行缩放,适合需要保持物理尺寸一致的项目(不同DPI下大小一致)
            - DPI: 每英寸像素点数
            - Physical Unit: 物理单位, 使用物理单位种类
            - Fallback Screen DPI: 备用屏幕DPI, 当无法获取设备DPI时使用的DPI值
            - Default Sprite DPI: 默认图片DPI, 用于计算UI控件大小
            - Reference Pixels Per Unit: 每单位参考像素, 多少个像素等于Unity一个单位,一般为100
            - 新单位参考像素 = 单位参考像素(Reference Pixels Per Unit)*Physical Unit / Default Sprite DPI
            - UI尺寸 = 图片像素大小 / (Pixels Per Unit / 新单位参考像素)
