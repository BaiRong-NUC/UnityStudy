1. 当图片资源是图集时,需要在设置时将模式设置为Multiple,可以使用Sprite Editor进行切割.

2. 参数说明:
- type: Automated: 自动选择模式; Grid By Cell Size: 按单元格大小划分; Grid By Cell Count: 按单元格数量划分.
- pivot: 设置图片的中心点位置,可以选择预设的位置或者自定义位置.
- custom pivot: 当pivot设置为Custom时,可以通过该参数自定义中心点的位置,范围是0-1之间的值,表示相对于图片宽高的比例.
- method: 如何处理现有对象
    - delete existing: 替换掉任意选择的任何矩形
    - smart: 智能处理现有对象,尽量保留现有对象同时创建新矩形.
    - safe: 添加新矩形不改变任何已经存在的矩形
- keep empty cells: 是否保留空白单元格.
- trim: 修剪每个单元格的透明像素边界.

3. Polygon多边形编辑(Sprite Mode)
- 如果使用的资源是多边形资源，可以在设置时将模式设置为Polygon,然后使用Sprite Editor中的Polygon Editor进行编辑.
- sides: 边数
- 注意: 从多边形切换到Single或Multiple模式时,需要重新生成Custom Outline
