1. SpriteAtlas:
    - 主要作用是: 将多个精灵图像打包到一个纹理图集中, 以减少渲染时的Draw Call数量, 提高性能.
    - 是Unity中的一种资源类型, 可以通过Sprite Atlas Asset创建和管理.
    - 注意: 使用Sprite Atlas后, 需要确保在代码或编辑器中正确引用精灵图像, 以便从图集中获取它们.
2. 使用: Edit -> Project Setting ->Editor -> Sprite Packer
- Mode
    - Enable For Build(Legacy Sprite Packer): Unity仅在构建时打包图集，在编辑模式下不会打包图集
    - Always Enabled(Legacy Sprite Packer): Unity在构建时打包图集，在编辑模式下运行前会打包图集
    - Enabled for Builds: Unity进在构建时打包图集，在编辑器模式下不会打包
    - Always Enabled: Unity在构建时打包图集，在编辑模式下运行前会打包图集
    - Disabled: 禁用图集打包功能
- Legacy Sprite Packer:传统打包模式,多了一个设置图片之间的间隔距离

3. 打图集面板相关参数
- project -> Assets ->右键 -> Create -> Sprite Atlas
- Type: 
    - Master: 主图集
        - Include in Build: 构建时包含图集
        - Allow Rotation: 允许图集中的精灵图像旋转以节省空间,UI图集一般不允许旋转
        - Tight Packing: 紧密打包精灵图像以节省空间,使用图片轮廓打包而不是图片矩形
        - Padding: 图集内精灵图像之间的间距,防止纹理边缘出现颜色混合
    - Variant: 变体图集,可以基于主图集创建不同分辨率的图集
        - Variant Scale: 设置变体图集的缩放因子（0~1），变体图集的大小是主图集乘以Scale的结果.
- 变体图集的主要作用是以主图集为基础，对它进行缩放产生一个新的图集副本。如果想使用变体图集中的内容，只需要勾选变体图集的Includeinbuild选项，而主图集禁用此选项即可
- Objects for Packing: 用于打包图集的对象列表.可以手动添加需要包含在图集中的精灵图像,或文件夹

4. 观察Draw Call数量
- Game窗口 -> Stats -> Batches
